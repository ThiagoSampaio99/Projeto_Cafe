### **2. Criação do S3 Bucket**

Criar um bucket novo:

Condições:

*   Bucket único (nome não existente na sua conta S3)
*   Pertencer a região (sa-east-1)


```python
import boto3

# Conexão com conta AWS
s3 = boto3.client('s3')

# Função para criar um S3 Bucket
def create_s3_bucket(bucket_name):

  try:
    response = s3.create_bucket(
        Bucket = bucket_name,
        CreateBucketConfiguration = {
            'LocationConstraint': 'sa-east-1'
        }
    )
    print(f'Bucket: "{bucket_name}" criado com sucesso!')

  except Exception as e:
    print(f'Erro ao criar o bucket: {e}')
```

### Obs na criação de new_bucket

*Erros principais:*
* Bucket já existente
* Bucket com nome inválido


*Nome de bucket (requisitos):*
* Somente deve conter:
1. Letras minúsculas
2. Números
3. Ponto final ( . )
4. Hífen ( - )
* Começar e terminar com letra ou número


https://docs.aws.amazon.com/AmazonS3/latest/userguide/bucketnamingrules.html?icmpid=docs_amazons3_console



```python
new_bucket = "conab-dados"
create_s3_bucket(new_bucket)
```
